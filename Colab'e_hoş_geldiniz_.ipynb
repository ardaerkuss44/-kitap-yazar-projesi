```
import pandas as pd
import os
from google.colab import files

# Önce dosyayı yükle
uploaded = files.upload()

# Yüklenen dosyaları kontrol et
print("Yüklenen dosyalar:", os.listdir())

# Dosya adını buradan doğru şekilde kopyala
df = pd.read_csv("TurkishBookDataSet.csv", on_bad_lines='skip', encoding='utf-8')

df.head()
```

```
# Kolonları incele
print("Kolonlar:", df.columns)

# İlk 10 kitabı göster
print(df[["name","author"]].head(10))

# Kaç farklı yazar var?
print("Toplam yazar sayısı:", df["author"].nunique())
```

```
book_name = input("Kitap ismini girin: ")

result = df[df["name"].str.lower() == book_name.lower()]

if not result.empty:
    print("Yazar:", result["author"].values[0])
else:
    print("Bu kitap veri setinde bulunamadı.")
```

```
if not result.empty:
    author = result["author"].values[0]
    print("Yazar:", author)
    print("Bu yazarın diğer kitapları:")
    print(df[df["author"] == author]["name"].head(5).to_list())
```
